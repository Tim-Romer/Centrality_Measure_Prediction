## Scale-Free Network:

```python
import networkx as nx
G = nx.scale_free_graph(100)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=False, node_size=50)
```

```python
print("Average Degree Centrality: " + str(round(sum(nx.degree_centrality(G).values())/len(G), 4)))
print("Average Closeness Centrality: " + str(round(sum(nx.closeness_centrality(G).values())/len(G), 4)))
print("Average Betweenness Centrality: " + str(round(sum(nx.betweenness_centrality(G).values())/len(G), 4)))
```

## Small-World Network:

```python
small_world = nx.watts_strogatz_graph(100, 3, 0.5)
```

```python
nx.draw_networkx(small_world, nx.spring_layout(small_world), with_labels=False, node_size=50)
```

```python
print("Average Degree Centrality: " + str(round(sum(nx.degree_centrality(small_world).values())/len(small_world), 4)))
print("Average Closeness Centrality: " + str(round(sum(nx.closeness_centrality(small_world).values())/len(small_world), 4)))
print("Average Betweenness Centrality: " + str(round(sum(nx.betweenness_centrality(small_world).values())/len(small_world), 4)))
```

## Random Network (Neither SF or SW):

The NetworkX library has a large library of graphs that can be accessed to generate randomly, which can be found here:
https://networkx.github.io/documentation/latest/reference/generators.html

We can generate a completely random graph, where the first argument is n and the second argument is m.

```python
random=nx.gnm_random_graph(100, 200)
%matplotlib inline
nx.draw(random, node_size=50)
```

```python
print("Average Degree Centrality: " + str(round(sum(nx.degree_centrality(random).values())/len(random), 4)))
print("Average Closeness Centrality: " + str(round(sum(nx.closeness_centrality(random).values())/len(random), 4)))
print("Average Betweenness Centrality: " + str(round(sum(nx.betweenness_centrality(random).values())/len(random), 4)))
```

## Scale-Free, Small-World Network (optional):

In theory, if we start with a Scale-Free network, and add some random paths between nodes, we can approach a Scale-Free, Small-World Network. This is because we will increase the clustering while maintaining the inverse power distribution of degrees.

```python
sf_sw = nx.scale_free_graph(100)
```

```python
nx.draw_networkx(sf_sw, nx.spring_layout(sf_sw), with_labels=False, node_size=50)
# initial scale-free network
```

```python
import random
for i in range(10):
    sample = random.sample(range(100), 2) # generates 2 unique random numbers from 0-99
    sf_sw.add_edge(sample[0], sample[1])
```

```python
nx.draw_networkx(sf_sw, nx.spring_layout(sf_sw), with_labels=False, node_size=50)
# scale-free network with random edges added
```

```python
print("Average Degree Centrality: " + str(round(sum(nx.degree_centrality(sf_sw).values())/len(sf_sw), 4)))
print("Average Closeness Centrality: " + str(round(sum(nx.closeness_centrality(sf_sw).values())/len(sf_sw), 4)))
print("Average Betweenness Centrality: " + str(round(sum(nx.betweenness_centrality(sf_sw).values())/len(sf_sw), 4)))
```
